```python
import re
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
url2 = 'https://es.wikipedia.org/wiki/Anexo:Asteroides_(1)-(1000)'
```

```python
req = requests.get(url2)
req
```

```python
contenido2 = req.content
```

```python
sopa2 = BeautifulSoup(contenido2, "html.parser")
sopa2
```

```python
tab = sopa2.findAll('tr')

tab
```

```python
type(tab)
```

```python
for i in tab:
    print(i.text)
```

```python
td = sopa2.findAll('td')
td
```

```python
body_td_txt = [str(i.text.replace('\n','')) for i in td]
body_td_txt
```

```python
cuerpo=[]
n=1
for i in body_td_txt:
    cuerpo.append(i.replace('('+str(n)+') ',''))
    if i.find('('+str(n)+')')!=-1:
        n= n+1

cuerpo
```

```python
cuerpo[:4]
```

```python
a = 0
lst = []
for i in range(0, len(cuerpo), 4):
    lst.append(cuerpo[a:i])
    a = i
```

```python
lst
```

```python
lst.pop(0)
```

```python
cabe=['Nombre',
 'Descubridor(es)',
 'Observatorio o Sitio de observación',
 'Fecha']
```

```python
df_plus = pd.DataFrame(lst, columns = cabe)
df_plus['id']=[i for i in range(1,1000)]
```

```python
df_plus = df_plus.reindex(columns=['id','Nombre',
 'Descubridor(es)',
 'Fecha',
 'Observatorio o Sitio de observación'])
df_plus
```
